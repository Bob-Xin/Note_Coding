### Hello PyMC

```python
fig, axes = plt.subplots(1,1,figsize=(15,4))

count_data = np.loadtxt('./000001SS.csv', delimiter=',')
count_data = count_data[:,1]
n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data, color='lightblue', alpha=0.3)
plt.ylabel("Closed Price")
plt.xlabel("Days")
plt.xlim(0, n_count_data);
```

```python
import pymc as pm

alpha = 1.0/count_data.mean()

lambda_1 = pm.Exponential("lambda_1", alpha)
lambda_2 = pm.Exponential("lambda_2", alpha)

tau = pm.DiscreteUniform("tau", lower=0, upper=n_count_data)
```

```python
print tau.random()
```

```python
@pm.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    out = np.zeros(n_count_data)
    out[0:tau] = lambda_1
    out[tau:] = lambda_2
    return out
```

```python
observation = pm.Poisson("obs", lambda_, value=count_data, observed=True)
```

```python
model = pm.Model([observation, lambda_1, lambda_2, tau])
```

```python
mcmc = pm.MCMC(model)
```

```python
mcmc.sample(40000, 10000)
```

```python
lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
tau_samples = mcmc.trace('tau')[:]
```

```python
fig = plt.figure(figsize=(15,5))

N = tau_samples.shape[0]

expected_texts_per_day = np.zeros(n_count_data)
for day in range(0, n_count_data):
    ix = day < tau_samples
    expected_texts_per_day[day] = (lambda_1_samples[ix].sum() + lambda_2_samples[~ix].sum()) / N

plt.plot(range(n_count_data), expected_texts_per_day, lw=4, color="#E24A33", label="Expected Price")
plt.bar(np.arange(len(count_data)), count_data, color="#348ABD", alpha=0.25, label="Actual Price")
plt.plot(count_data)

plt.legend(loc="upper left")
print expected_texts_per_day[0], expected_texts_per_day[-1]
```

```python
import pymc as pm

lambda_1 = pm.Exponential("lambda_1", alpha)
lambda_2 = pm.Exponential("lambda_2", alpha)
lambda_3 = pm.Exponential("lambda_3", alpha)

tau_1 = pm.DiscreteUniform("tau_1", lower=0, upper=n_count_data-1)
tau_2 = pm.DiscreteUniform("tau_2", lower=tau_1, upper=n_count_data)
```

```python
@pm.deterministic
def lambda_(tau_1=tau_1, tau_2=tau_2, lambda_1=lambda_1, lambda_2=lambda_2, lambda_3=lambda_3):
    out = np.zeros(n_count_data)
    out[0:tau_1] = lambda_1
    out[tau_1:tau_2] = lambda_2
    out[tau_2:] = lambda_3
    return out
```

```python
observation = pm.Poisson("obs", lambda_, value=count_data, observed=True)
model = pm.Model([observation, lambda_1, lambda_2, lambda_3, tau_1, tau_2])
```

```python
mcmc = pm.MCMC(model)
mcmc.sample(40000, 10000)
```

```python
lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
lambda_3_samples = mcmc.trace('lambda_3')[:]
tau_1_samples = mcmc.trace('tau_1')[:]
tau_2_samples = mcmc.trace('tau_2')[:]
```
